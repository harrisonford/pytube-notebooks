# pytube: Loading them videos

pytube is a library that can be loading as simple as using pip,
this is because it was built dependency-free!

```python
from pytube import YouTube, Playlist

print('done!')
```

## Downloading a video

```python
# this is one way of doing it
short_url = 'https://youtu.be/2lAe1cqCOXo'
YouTube(short_url).streams.first().download(output_path="./tmp" ,filename='short_test')
```

```python
# we can also give the full extension
long_url = 'http://youtube.com/watch?v=2lAe1cqCOXo'
yt = YouTube(long_url)
yt.streams\
    .filter(progressive=True, file_extension='mp4')\
    .order_by('resolution')\
    .desc()\
    .first()\
    .download(output_path="./tmp" ,filename='long_test')
```

## Manipulate captions

```python
# get video object
yt = YouTube('https://www.youtube.com/watch?v=cfsiVQDL6zg')
# list captions
print(yt.captions.all())
```

```python
# get a caption by language code
caption = yt.captions.get_by_language_code('a.es')
# print it
print(caption.generate_srt_captions())
```

## Putting all together
We will download a playlist and save the 'a.es' captions.

```python
# Criticas QLS playlist
pl = Playlist("https://www.youtube.com/watch?v=cfsiVQDL6zg&list=PLt5vaMQT6yBGGTJcbvGgtz4Y5i9FF7OtV")
for video in pl.videos:
    video.streams.first().download(output_path="./tmp" ,filename=video.title)
    # save the caption
    caption = video.captions.get_by_language_code('a.es')
    with open(video.title + ".txt", "w") as text_file:
        text_file.write(caption.generate_srt_captions())
```
